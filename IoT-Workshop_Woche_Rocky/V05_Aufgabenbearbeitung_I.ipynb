```python
import pandas as pd  #Einfügen der Pandas-Bibliothek über die Abkürzung "pd"
import datetime      #um die Datentypen Datum und Zeit nutzen zu können
```

```python
df = pd.read_csv("C:/Users/Rocky Telatko/Documents/RT/Python/IoT-Workshop_Woche/ProdTag_12_05_2021.csv", 
                 sep = ";")   #Einlesen der csv-Daten
df.head(n=10)   #Ausgabe der ersten 10 Zeilen des Datensatzes
```

```python
df.info()   #Ausgabe der Datensatzstruktur
```

```python
df["Zykluszeit"] = pd.to_datetime(df["Endzeit"]) - pd.to_datetime(df["Startzeit"])   #Berechnung der Zykluszeit je Prozessschritt
df["Zykluszeit"] = df["Zykluszeit"].dt.total_seconds()   #Umwandlung der Zykluszeit in Sekunden
df.head(n=5)   #Ausgabe der ersten 5 Zeilen des Datensatzes
```

```python
df.dtypes   #Ausgabe der Datentypen je Spalte des Datensatzes
```

```python
#Transportzeit und Zykluszeit je Arbeitsstation (summiert)
selectAS = df[["Arbeitsstation", "Transportzeit", "Zykluszeit"]]   #Selektieren des Datensatzes nach den beschriebenen Spalten
prozessDauernAS = selectAS.groupby("Arbeitsstation").sum()   #Gruppierung nach den Arbeitsstationen und Aufsummieren der Werte
prozessDauernAS["Gesamtzeit"] = prozessDauernAS["Transportzeit"] + prozessDauernAS["Zykluszeit"]   #Berechnung der Gesamtzeit in extra Spalte
prozessDauernAS   #Ergebnisausgabe
```

```python
#NaN werden mitunter eigenständig gefiltert: "Operations in general exclude missing data" (das muss aber nicht sein!)
selectAS2 = selectAS.dropna()   #Entfernen von Zeilen mit fehlenden Werten
prozessAnz1 = selectAS.groupby("Arbeitsstation").count()   #Gruppierung nach Arbeitsstationen und Zählen der eingetragenen Werte je Station
prozessAnz2 = selectAS2.groupby("Arbeitsstation").count()   #Gruppierung nach Arbeitsstationen und Zählen der eingetragenen Werte je Station
print(prozessAnz1)   #Ergebnisausgabe
print(prozessAnz2)   #Ergebnisausgabe
#Anhand der Transportzeit ist die Filterung der NaN-Werte zu erkennen, da die Zykluszeit unter Ausschluss der NaN-Werte berechnet wurde 
```

```python
meanProzessDauerAS = selectAS.groupby("Arbeitsstation").mean()   #Gruppierung nach den Arbeitsstationen und ermitteln des Durchschnitts über die Werte
meanProzessDauerAS["Gesamtzeit"] = meanProzessDauerAS["Transportzeit"] + meanProzessDauerAS["Zykluszeit"]   #Ermitteln der Gesamtzeit in extra Spalte
meanProzessDauerAS   #Ergebnisausgabe
```

```python
#Durchschnittliche Transportzeit und Zykluszeit je Arbeitsstation und Produktvariante
selectAS_PV = df[["Arbeitsstation", "Produktvariante", "Transportzeit", "Zykluszeit"]]   #Selektieren des Datensatzes nach den beschriebenen Spalten
meanProzessDauerAS_PV = selectAS_PV.groupby(["Arbeitsstation", "Produktvariante"]).mean()   ##Gruppierung nach den beschriebenen Spalten und ermitteln des Durchschnitts über die Werte
meanProzessDauerAS_PV["Gesamtzeit"] = meanProzessDauerAS_PV["Transportzeit"] + meanProzessDauerAS_PV["Zykluszeit"]   #Ermitteln der Gesamtzeit in extra Spalte
meanProzessDauerAS_PV   #Ergebnisausgabe
```

#### Auswertung der statistischen Werte

Anhand der Anzahl der Transporte (bei AS 5 jeweils zwei Prozessschritte je Werkstück) und der Anzahl der Zykluszeiten lassen sich Rückschlüsse ziehen, an welcher Stelle es Probleme (Störungen) gab --> bei Werkstück 8010 und 8111 bei bzw. nach AS 8, bei Werkstück 8011 bei bzw. nach AS 7 (Abfolge AS 5-11-8-6-7-5).

Die Standardabweichung bei der Transportzeit der AS 5 ist ungewöhnlich. Die Ursache ist, dass die erste erkannte Transportzeit parallel zum ersten Prozessschritt erfasst wird und damit, im Gegensatz zum zweiten Prozessschritt der AS, einen Nullwert darstellt.

Der Maximalwert der Zykluszeit bei AS 7 und Werkstück 8011 (siehe auch die hohe Standardabweichung), neben dem von AS 6 bei Werkstück 8010, signalisiert ebenfalls ein ungeplantes Ereignis bzw. ungeplante Ereignisse. Um zu prüfen, ob es sich um ein Ereignis oder mehrere handelt, wird eine Filterung durchgeführt (siehe Befehl unten).

Es ist anhand der Standardabweichung erkennbar, dass die Engpassstation AS 8 den instabilsten Prozess darstellt --> woran liegt das? Zur Beantwortung der Frage oder dem Erarbeiten weiterer Fragen wären weitere Daten hilfreich!
Ebenfalls auffällig sind die Minimalwerte bei AS 8 bei Werkstück 8010 und 8111, welche ebenfalls auf ungeplante Ereignisse (Störungen) hinweisen (siehe Befehl unten).

Der Maximalwert der Zykluszeit bei AS 11 und Werkstück 8011 signalisiert ein weiteres ungeplantes Ereignis bzw. ungeplante Ereignisse. Um zu prüfen, ob es sich um ein Ereignis oder mehrere handelt, wird eine Filterung durchgeführt (siehe Befehl unten).

**Fazit: Fokus muss auf AS 8 gelegt werden!**

```python
df.loc[(df['Arbeitsstation'] == 7) & (df['Zykluszeit'] >= 10.0)]   #Ausgabe der Zeilen mit überdurchschnittlichen Werten bei der Zykluszeit von Station 7
```

wahrscheinliche Ursache bei 8011: Defekt

```python
df.loc[(df['Arbeitsstation'] == 11) & (df['Zykluszeit'] >= 5.0)]   #Ausgabe der Zeilen mit überdurchschnittlichen Werten bei der Zykluszeit von Station 11
```

wahrscheinliche Ursache: fehlendes Material oder Defekt

```python
'''
Prüfen, wie sehr die Zeiten der Produktvariante 8010 an Arbeitsstation 8 schwankt - sind einzelne oder mehrere Ereignisse betroffen?
Werkstück 8010 hat Durchschnittswert bei 106 Sekunden (Median = 103s)
'''
df.loc[(df['Arbeitsstation'] == 8) & (df['Produktvariante'] == 8010) & (df['Zykluszeit'] >= 110.0)]   #Ausgabe der Zeilen mit überdurchschnittlichen Werten bei der Zykluszeit von Station 8 bei Produktvariante 8010
```

Das Ergebnis verdeutlicht den Handlungsbedarf aufgrund der Instabilität der Vorgänge!

```python
df.loc[(df['Arbeitsstation'] == 8) & (df['Produktvariante'] == 8010) & (df['Zykluszeit'] <= 100.0)]   #Ausgabe der Zeilen mit unterdurchschnittlichen Werten bei der Zykluszeit von Station 8 für Produktvariante 8010
```

Vermutlich handelte es sich hier um eine Störung, wodurch eine Bearbeitung unterbrochen und als zwei separate Bearbeitungen erfasst wurde.
