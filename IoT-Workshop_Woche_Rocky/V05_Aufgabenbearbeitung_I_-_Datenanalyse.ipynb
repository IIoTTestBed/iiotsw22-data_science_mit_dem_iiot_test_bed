```python
#df = pd.read_csv("euerPfad/Dateiname.csv", sep = ";") --> für Workshop
df = pd.read_csv("C:/Users/Rocky Telatko/Documents/RT/Python/IoT-Workshop_Woche/ProdTag_12_05_2021.csv", 
                 sep = ";")   #Einlesen der csv-Daten
df.head(n=10)   #Ausgabe der ersten 10 Zeilen des Datensatzes
```

```python
df["Zykluszeit"] = pd.to_datetime(df["Endzeit"]) - pd.to_datetime(df["Startzeit"])   #Berechnung der Zykluszeit je Prozessschritt
df["Zykluszeit"] = df["Zykluszeit"].dt.total_seconds()   #Umwandlung der Zykluszeit in Sekunden
```

```python
#Berechnung der Transportzeit für die ersten 1091 Zeilen, da NaN-Werte ab Zeile 1095 (PS = 40) beginnen
import datetime      #um die Datentypen Datum und Zeit nutzen zu können

dfShort = df.iloc[:1091]   #Filtern nach den ersten 1091 Zeilen
dfShort = dfShort.copy()   #Erstellen einer Kopie des Datensatzes
#dfShort.info()

zeile = 0
counter = 0
for row in range(0, len(dfShort)):   #zeilenweises Durchgehen des Datensatzes 
    if dfShort.iloc[zeile]['Prozessschritt'] != 10:   #Wenn der Prozessschritt ungleich 10 ist, berechne die Transportzeit und füge sie der neuen Spalte hinzu
        vorherigeZeile = zeile - 1
        transportZeit = pd.to_datetime(dfShort.iloc[zeile]["Startzeit"]) - pd.to_datetime(dfShort.iloc[vorherigeZeile]["Endzeit"])
        dfShort.at[zeile, "TransportzeitKorrekt"] = transportZeit
        zeile += 1
    else:   #Wenn der Prozessschritt gleich 10 ist, setze die Transportzeit auf 0 und füge sie der neuen Spalte hinzu
        nullWert = pd.to_datetime(dfShort.iloc[0]["Startzeit"]) - pd.to_datetime(dfShort.iloc[0]["Startzeit"])
        dfShort.at[zeile, "TransportzeitKorrekt"] = nullWert
        zeile += 1

dfShort["TransportzeitKorrekt"] = dfShort["TransportzeitKorrekt"].dt.total_seconds()   #Umwandlung der neuen Transportzeit in Sekunden
dfShort.head(10)
#df.info()
#dfShort.iloc[-1:] #Ausgeben der letzten Zeile des Datensatzes
```
